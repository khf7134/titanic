```python
datadir_path = "../data/"
outputdir_path = "../output/"

train_data = pd.read_csv(datadir_path + "train.csv")
train_data.head()
```

```python
test_data = pd.read_csv(datadir_path + "test.csv")
test_data.head()
```

```python
"""
pd.read_csv(file_name)の機能
カンマ区切りデータを読みたい場合はread_csv,/t区切りデータの場合はread_table,それ以外の場合は引数指定
headerがないデータの場合、勝手に1行目がヘッダーになっちゃう
指定する方法によって、headerにならなかったり、headerを仮指定できる。
インデックスも指定しないと認識されない。
usecolsで指定することによって、特定の列のみ抽出可能、この時特定の列以外を読ませる時、usecols=lambda x: x is not 'b'という形で使う
データの指定もできるお、ただ指定しても欠損値はfloat処理
‘#N/A’, ‘#N/A N/A’, ‘#NA’, ‘-1.#IND’, ‘-1.#QNAN’, ‘-NaN’, ‘-nan’, ‘1.#IND’, ‘1.#QNAN’, ‘N/A’, ‘NA’, ‘NULL’, ‘NaN’, ‘n/a’, ‘nan’, ‘null’は全てNaN処理
NaNは追加もでき、指定した値のみNaNにするものもある。
"""
```

```python
"""
data_
survival=生きていたか_0死,1生
pclass=チケットクラス_11st,22nd,33rd
sex=sex
age=age
sibsp=配偶者、兄弟が乗っていたか
parch=親や子供が乗っていたか(乳母の場合は0)
ticketnumber=ticketnumber
fare=passengerfare
cabin=cabinnumber
train.csvのデータ
passengerid,survived(0or1),Pclass,Name,Sex,Age,Sibsp,Parch,Ticket,Fare
test.csvのデータ
Survivedなし、最後にcabinが入っている。
"""

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch",]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

"""
n_estimators=number_of_tree
max_depth=maximum_depth_of_the_tree
random_state=seeds
"""
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=1)
model.fit(X, y)
"""
これ、model.fitってXのNaN扱いどうしてんの
n_estimatorsはデータセットの数
max_samplesという引数で抽出する際のデータの上限
max_featuresは抽出する際の特徴量(ID)の指定
NaNの扱いはまた明日。
"""
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv(outputdir_path+'decision_tree1_1.csv', index=False)
print("Your submission was successfully saved!")
```
